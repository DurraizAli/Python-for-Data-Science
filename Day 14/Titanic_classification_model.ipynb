# Classification model of Titanic

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score
from sklearn import tree
```

```python
df= sns.load_dataset('titanic')
```

```python
X = df[['pclass','sex','age','sibsp','parch','fare']]
y= df[['survived']]
```

```python
X = pd.get_dummies(X,columns=['sex'])
```

```python
# filling null values with mean of age
X['age'].fillna(X['age'].mean(), inplace=True)
# filling null values with mean of fare
X['fare'].fillna(X['fare'].mean(), inplace=True)
```

```python
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
prd=model.predict(X_test)
```

```python
print("The Accuracy score is : ",accuracy_score(y_test,prd))
print("F1 Accuracy score is : ",f1_score(y_test,prd))
print("The Precision Score is : ",precision_score(y_test,prd))
```

```python
tree.plot_tree(model.fit(X,y),filled=True)
plt.savefig("classification.svg")
plt.show()
```
