# Linear Regression


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df= sns.load_dataset('titanic')
```

We will distinguish the X and y variables


```python
df.age.fillna(value=df['age'].mean(),inplace=True)
```

```python
df.fare.fillna(value=df['fare'].mean(),inplace=True)
```

```python
df_linear = df[['age','fare']]
#df_loci=df.iloc[:,3:5] -> how to use iloc method
```

```python
sns.lmplot(x='age',y='fare',data=df_linear)
```

Removing Data

```python
sns.boxplot(df_linear['age'])
```

```python
df_linear=df_linear[(df_linear['age']>5)& (df_linear['age']<55)]
```

```python
sns.boxplot(df_linear['fare'])
```

```python
df_linear
```

```python
df_linear=df_linear[(df_linear['fare']>5)& (df_linear['fare']<50)]
```

```python
df_linear['fare'].shape
```

```python
df_linear['age'].shape
```

```python
X=df_linear[['age']]
y=df_linear['fare']
```

```python
model=LinearRegression()
```

Splitting data using train test split

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8)
```

```python
model.fit(X_train,y_train)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
prd=model.predict(X_test)
```

```python
sns.scatterplot(x=y_test,y=prd)
```
