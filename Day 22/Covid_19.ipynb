```python
local=df[df['Travel_history']=='Local - Social Contact'].count()
travelled=df[df['Travel_history']!='Local - Social Contact'].count()
```

```python
local=df[df['Travel_history']=='Local - Social Contact'].count().shape[0]
travelled=df[df['Travel_history']!='Local - Social Contact'].count().shape[0]
labels=['locals','travelers']
values=[local,travelled]
fig=px.pie(df,values=values,names=labels,title='Difference of people who travelled and got Covid vs locals',hole=0.5)
fig.show()
```
